```python
import psycopg2
```

```python
# STEP 1: Connect to default Postgres database
```

```python
try:
    # Establish connection to the default 'postgres' database
    conn = psycopg2.connect("host = 127.0.0.1 dbname=postgres user=postgres password=Pavan@2712")
    print("Connected Successfully")
except psycopg2.Error as e:
    print("Error: Could not make connection to Postgres database")
    print(e)
```

```python
# STEP 2: Create cursor object
```

```python
try:
     # Cursor is used to execute SQL queries
    cur = conn.cursor()
except psycopg2.Error as e:
    print("Error: Could not get cursor to the database")
    print(e)
```

```python
# Enable autocommit to execute CREATE DATABASE
conn.set_session(autocommit=True)
```

```python
# STEP 3: Create a new database
```

```python
try:
    cur.execute("Create database myfirstdb")
except psycopg2.Error as e:
    print(e)
```

```python
# Close connection to default database
try:
    conn.close()
except psycopg2.Error as e:
    print(e)
```

```python
# STEP 4: Connect to newly created database
```

```python
try:
    conn = psycopg2.connect("host=127.0.0.1 dbname=myfirstdb user=postgres password=Pavan@2712")
    print("Connected to myfirstdb")
except psycopg2.Error as e:
    print("Error: Could not make conncetion to the postgres database")
    print(e)

# Create cursor for new database
try:
    cur = conn.cursor()
except psycopg2.Error as e:
    print("Error: Could not get cursor to the database")
    print(e)

# Enable autocommit for table creation and inserts
conn.set_session(autocommit=True)
```

```python
# STEP 5: Create students table
```

```python
try:
    cur.execute("CREATE TABLE IF NOT EXISTS students (student_id int, name varchar, age int, gender varchar, subject varchar, marks int);")
    print("Student TABLE created")
except psycopg2.Error as e:
    print("Error: Issue creating table")
    print(e)
```

```python
# STEP 6: Insert records into table
```

```python
try:
    cur.execute("INSERT INTO students (student_id, name, age, gender, subject, marks) VALUES (%s, %s, %s, %s, %s, %s)", (1, "Yug", 23, "Male", "Python", 70))
except psycopg2.Error as e:
    print("Error: Inserting rows")
    print(e)

try:
    cur.execute("INSERT INTO students (student_id, name, age, gender, subject, marks) VALUES (%s, %s, %s, %s, %s, %s)", (2, "Rishi", 21, "Male", "ReactJS", 60))
except psycopg2.Error as e:
    print("Error: Inserting rows")
    print(e)
```

```python
# STEP 7: Fetch and display records
```

```python
try:
    cur.execute("SELECT * FROM students;")
except psy.Error as e:
    print("Error: SELECT *")
    print(e)

# Fetch rows one by one
row = cur.fetchone()
while row:
    print(row)
    row = cur.fetchone()
```

```python
# STEP 8: Close cursor and connection
```

```python
cur.close()
conn.close()
```
